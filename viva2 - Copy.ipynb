```python
#Q1--
import numpy as np
import matplotlib.pyplot as plt

plt.title("bar chart of function x=y2",loc='center',color='green')
plt.xlabel("number(y)")
plt.ylabel("square of the number(x)")
x = np.arange(15)
y = x**2
plt.xticks(np.arange(15))
plt.bar(x,y,width=0.8)
plt.show()
```

```python
#Q2--
import numpy as np
angles = np.array([0,30,45,60,90])
sines = np.sin(np.deg2rad(angles))
print("sine: array of angles given in degree",sines)

coses = np.cos(np.deg2rad(angles))
print("cosine: array of angles given in degree",coses)

tans = np.tan(np.deg2rad(angles))
print("tangent: array of angles given in degree", tans)
```
