```python
def fact(a):
  if a==0 or a==1:
    return 1
  else:
    ans=1
    i=1
    while i<=a:
      ans*=i
      i+=1
    return ans

print(fact(6))
```

```python
def student_info(*args,**kwargs):   #it is used when we don't know how many arguments are going to be passed
  print(args)   #args returns the simple arguments 
  print(kwargs)     #kwargs returns the key value arguments
#here math and art ar eour positional arguments
#name and age are key word argument
student_info('math','art',name='Bhanu',age=16)
```

```python
courses=['math','physics','chemistry']
info={'name':'Bhanu','age':16}

student_info(courses,info)    #this function is passing all the list and the dictionary as positional arguments only 
```

```python
#to pass list as the positional and the dictionary as the keyword arguments we should use a star and double star
student_info(*courses,**info)
```
