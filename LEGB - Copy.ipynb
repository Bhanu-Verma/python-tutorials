```python
#LEGB---- local, enclosed, global, built-in

#example of local variable--

def p():
  x=20    #this x is a local variable
          #its scope is in the function where it is defined
  print("x inside the function is ",x)

x=10
print("x outside the function is ",x)
p()
```

```python
# if we define a variable outside the function its scope is global and we can access it anywhere in our program
x=100   #here this is global variable
        #its scope is in the complete code 
def p():
  x=20      #this is a local variable
  print("x inside the function is:",x)
p()
print("x outside the function is:",x)
```

```python
#suppose in our above mentioned function we want to access the global x inside the function then we'll have to use global keyword
x=100   #here this is global variable
def plus20():
  global x
  x+=20      #this is a not local variable
  print("x inside the function is:",x)
  x=10     #now we can change the value of x also
  print("x inside the function is:",x)
plus20()
print("x outside the function is:",x)   #print is a built-in method we can use it anywhere in python environment
```

```python
#NOW LET'S TALK ABOUT NON-LOCAL OR ENCLOSED VARIABLES
#this concept is mainly used when we have nested functions

x=100     #this x is global
def outer():
  x=50      #after defining the inner function this is neither a local nor a global variable
            #now it is termed as a non local variable
  def inner():
   
    x=10    #now this is local variable for this function
```

```python
#if we want to access the nonlocal variable in the inner function--
x=100     #this x is global
def outer():
  x=50      #after defining the inner function this is neither a local nor a global variable
            #now it is termed as a non local variable
  def inner():
    nonlocal x
    x=10 
  inner()  
  print("value of x after the inner function call ",x)

outer()
```

```python
#LEGB RULE -- this rule says that when we try to access a variable it is first searched in the local scope, then in the enclosed scope
#,global scope and after that it is searched in the built-in scope.
#NOTE-- here it's not necessary that search will alwaus begin with the local scope it depends from where we are accessing the variale

#now you can easily understand this example--
#spam = "abc"
def scope_test():
    def do_local():
        spam = "local spam"

    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"

    def do_global():
        global spam
        spam = "global spam"

    spam = "test spam"
    do_local()
    print("After local assignment:", spam)
    do_nonlocal()
    print("After nonlocal assignment:", spam)
    do_global()
    print("After global assignment:", spam)

scope_test()
print("In global scope:", spam)

```
