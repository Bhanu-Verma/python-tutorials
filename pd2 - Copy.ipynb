```python
#adding a column to a data frame
dict2 = {
    'name':['Bhanu','RAmu','Vishnu'],
    'class':['B tech' for i in range(3)],
    'roll':[21,27,94]
}
df2 = pd.DataFrame(dict2)

df2
```

```python
#Adding a new column 
df2['doubleName']=df2['name']*2
df2
```

```python
#drop columns and rows--
df2.drop('class',axis=1,inplace=True)
```

```python
df2.drop(1,axis=0,inplace=True)
df2
```
