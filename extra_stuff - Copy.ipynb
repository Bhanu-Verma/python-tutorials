```python
#functools is used to implement dp in python--
import functools as fc
import time

@fc.lru_cache(maxsize=None)
def fx(n):
    time.sleep(5)
    return n*5

print(fx(20))
print("done for 20")
print(fx(2))
print("done for 20")
print(fx(6))
print("done for 20")
print(fx(20))
print("done for 20")
print(fx(2))
print("done for 20")
print(fx(6))
print("done for 20")
```
