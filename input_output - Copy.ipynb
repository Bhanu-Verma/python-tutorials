```python
#REPR AND STR FUNCTIONS

#In Python, the repr() function returns a string representing a printable version of an object
#Note that the repr() function is different from the str() function, which returns a string representing a human-readable version of an object.

a=[1,2,3,4,5]
print(repr(a))
print(str(a))
```

```python
class Person:
  def __init__(self, name, age):
      self.name = name
      self.age = age

  def __repr__(self):
      return f'Person(name={self.name!r}, age={self.age!r})'
  
  def __str__(self):
      return f'{self.name} ({self.age})'

p = Person('Alice', 30)

print(repr(p)) # Output: Person(name='Alice', age=30)
print(str(p)) # Output: Alice (30)
```

```python
#USING FORMAT---
a=2
b=5
c=a+b
print("the sum of {0} and {1} is {2}".format(a,b,c))
print("the sum of {} and {} is {}".format(a,b,c))
print("the sum of {first_val} and {second_val} is {result} ".format(first_val=a,
                                                                     second_val=b,
                                                                     result=c))
#position and keyword arguments can be mixed just randomly
print("the sum of {0} and {1} is {ans}".format(a,b,ans=c))
#we can make a dictionary of all the values and pass it as an argument to the format function
#here we pass dictionary as a positional argument
dic={'num1':100,'num2':2,'sum':3}
print("the sum of {0[num1]:d} and {0[num2]:d} is {0[sum]}".format(dic))

#if we pass the dictionary as a keyword argument we needn't put that zero
print("the sum of {num1:1d} and {num2:5d} is {sum:10d}".format(**dic))

#if you are thinking what is this d doing here--
#     it will limit the code to integers only
#     the number written before d determines the fields we are going to use for output
```
