```python
#decorators in python
#decorator is a function that returns another function


#need of decorators -- supppose we want to print a good mornging statement before every function of our module
#in this kind of cases we use decorators


def greet(fx):
  def mfx(*args,**kwargs):    #if the function has arguments
    print("Good morning")
    fx(*args,**kwargs)
    print("thanks for using this function")
  return mfx
```

```python

def add(a,b):
  print(a+b)

greet(add)(1,2)
```

```python
@greet
def addd(n,m):
  ans = n+m
  return ans

x=addd(2,3)
print(x)    #we are getting none value because it is returning the result to the decorator only
```

```python
#now listen carefully getters are not like c++. here we are defining any function in the class and we want that the value this function returns can be used in a normal way we can do this with the help of getter for this we use property named built-in fuction decorator


class complex:
  def __init__(self,x,y):
    self.r=x
    self.i=y
    return
  @property
  def magnitude(self):
    return math.sqrt(self.r**2+self.i**2)

a=complex(1,2)
#print(f"magnitude of a is",a.magnitude())   #now this gives an error because after writing @property function behaves as a simple attribute
print(f"magnitude of a is",a.magnitude)   #this will work well
x=a.magnitude   #now it is not refrence to that function
print(x)

```

```python
#USING THIS SYNTAX WE CAN ACCESS THE RESULT OF FUNCTION METHODS BUT WE CAN'T CHANGE THE VARIABLES VALUES
#to do that we use setters
class complex:
  def __init__(self,x,y):
    self.r=x
    self.i=y
    return
  @property
  def magnitude(self):
    return math.sqrt(self.r**2+self.i**2)
  @property
  def give_r(self):
    return self.r
  @give_r.setter
  def r(self,nr):
    if(nr<0):
      raise ValueError('WTF')
    else:
      self.r = nr
    
a=complex(1,2)
a.r = 4
print("values of real parts are: ",a.give_r)
```

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise ValueError("Radius must be a positive value.")
        self._radius = value

# Create an instance of the Circle class
circle = Circle(5)

# Get the current radius
print(circle.radius)  # Output: 5

# Update the radius using the setter
circle.radius = 10

# Get the updated radius
print(circle.radius)  # Output: 10

# Try to set an invalid value (negative radius)
circle.radius = -3 
```

```python
#Derived classes may override methods of their base classes
#Because methods have no special privileges when calling other methods of the same object, a method of a base class that calls another method defined in the same base class may end up calling a method of a derived class that overrides it.


#Method overriding is a feature in object-oriented programming that allows a subclass to provide its own implementation for a method that is already defined in its superclass.

#In Python, method overriding occurs when a subclass defines a method with the same name as a method in its superclass. When an object of the subclass is created, the method in the subclass is called instead of the method in the superclass.

class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("The dog barks")

animal = Animal()
animal.speak()  # Output: The animal makes a sound

dog = Dog()
dog.speak()  # Output: The dog barks
```

```python
class Reverse:
    """Iterator for looping over a sequence backwards."""
    def __init__(self, data):
        self.data = data
        self.index = len(data)

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == 0:
            raise StopIteration
        self.index = self.index - 1
        return self.data[self.index]

s=Reverse([1,2,3,4,5])
for x in s:
  print(x)
```

```python
class two_step_iter:
  def __init__(self,data):
    self.data = data
    self.limit = len(data) 
  
  def __iter__(self):
    self.index = -2
    return self

  def __next__(self):
    if self.index >= self.limit-2:
      raise StopIteration
    else:
      self.index += 2
      return self.data[self.index]


a = two_step_iter([1,2,3,4,5,6,7,8])
for i in a:
  print(i)
```

```python
#GENERATORS---

#generators are written like usual function but they use yield statement to return whenever they want to return
#generators are used because now we needn't store the data in our memory we can make it on the fly
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        yield data[index]

for item in reverse([1,2,3,4,5]):
  print(item)
```

```python
#In Python, a static method is a type of method that belongs to a class rather than an instance of the class. Static methods do not operate on instance-specific data and do not have access to the instance itself. They are defined using the @staticmethod decorator


class Math:
    def __init__(self,num) -> None:
        self.num = num

    def addtonum(self,n):
        self.num = self.num + n
    @staticmethod
    def add2(a,b):  #self argument is not required at all
        return a+b
    
a = Math(5)
a.addtonum(6)
print(a.num)
Math.add2(1,3)      #can also be accessed from class name
a.add2(1,2)
```

```python
#class methods in python
class employee:
    company = "google"
    def __init__(self,name,salary) -> None:
        self.name = name
        self.salary = salary
    @classmethod
    def changeCompany(cls, newcmp):
        cls.company=newcmp
        
e1 = employee("bhanu",1200)
e1.changeCompany("microsoft")      #now it will really change the company

print(employee.company)
```

```python
#using class methods as constructors--
#now suppose in above class we are going to get a string like "bhane-100" then we have to first parse the data and then send to constructor


class employee:
    company = "google"
    def __init__(self,name,salary) -> None:
        self.name = name
        self.salary = salary
    @classmethod
    def fromStr(cls, string):
        return cls(string.split("-")[0],string.split("-")[1])
        
e1 = employee("bhanu",1200)
string = "Bhanu2-1000"
e2 = employee.fromStr(string)
print(e2.name,e2.salary)
```

```python
#operator overloading--

class complex:
    def __init__(self,x,y):
        self.r=x
        self.i=y
        return
  
    def __str__(self) -> str:     #dunder method
        return f"{self.r} + {self.i}i"

    def __add__(self,x):
       return complex(self.r+x.r, self.i+x.i)
 
    @property
    def magnitude(self):
        return math.sqrt(self.r**2+self.i**2)
  

identity = complex(0,1)
print(identity)

one = complex(1,0)

print(one+identity)
```

```python
#multiple inheritance--
class person:
    def __init__(self, name):
        self.name = name
    

class male:
    def __init__(self, gender) -> None:
        self.gender = gender
    

class student(person,male): #multiple inheritance
    def __init__(self, name, gender):
        super().__init__(name)
        super().__init__(gender)
    def say(self):
        print("I am both person and male")

#NOTE == if we define two same methods in the both parent calsses and call that method, then the method from the class that we wrote first while inheriting is called first

bhanu = student("bhanu","male")
print(bhanu.name)   #output male because first method is called both of times

```

```python
#multilevel inheritance--

class my_nano(nano):
    def __init__(self,dist):
        self.dist = dist
    def show_dist(self):
        print(f"{self.dist}")

x = my_nano(1000)
x.show_dist()
x.run()
```
