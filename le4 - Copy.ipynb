```python
#TUPLES
#tuples are immutable
#tuples are also a very imp data type but there is a major difference between tuples and lists that tuples can't be modified
tuple1=(1,2,3,4,5)
tuple2=tuple1
#tuple1[0]=99      #it is an error because it doesn't support assignment
print(tuple1)
print(tuple2)

i=0
while i<5:
  print(tuple1[i])
  i+=1
```

```python
#SET
#in a set we can't access its elements randomly because they are not in a order they are random
set1={1,2,3,45}
print(set1)
#set doen't contain any duplicates
set2={1,1,2,3,4,1,2}
print(set2)
print(1 in set1)
```

```python
print(set1.intersection(set2))    #it is used to get the intersection of two sets
print(set1.difference(set2))      #it gives all the courses of the set1 that are not in set2
print(set1.union(set2))     #it returns the union
```
