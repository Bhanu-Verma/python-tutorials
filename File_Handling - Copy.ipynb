```python
#this returns a file object, it has three arguments -- two positional, one keyword
#first argument is a string containing name of the file
#sencond is also a string spedifying the format
#third is the encoding scheme

f = open('/content/crtl_shift_skill.txt','r',encoding="utf-8")
print(f.read())
f.close()
print(f.closed)   #it returns a boolean value indicating that file has been closed or not

```

```python
#WITH KEYWORD
#here we need to close the file manually after its work is finished but if we use with 
#it is close itself evenif any error occurs, we needn't do it

with open('/content/crtl_shift_skill.txt','r') as f:
  read_data = f.read()

print(read_data)
print(f.closed)
```

```python
#METHODS OF FILE OBJECTS--

#1) f.read(size)
# size tells how much bytes of the file we want to be read if it is not specified or it 
# is set to zero it reads complete file 
with open('/content/crtl_shift_skill.txt','r') as f:
  read_data=f.read(10)
print(read_data)
```

```python
#2) f.readline()
# this returns the first line of the file when called for the first time
# when used second time is returns second line
with open('/content/crtl_shift_skill.txt','r') as f:
  s1=f.readline() #it returns line with the \n
  s2=f.readline()
  s3=f.readline()
  s4=f.readline()
print(s1,s2,s3,s4)


# we can use looping to read the lines of file::
with open('/content/crtl_shift_skill.txt','r') as f:  
  for line in f:
    print(line,end='') #if you don't specify end it will show two \n---one of line another of print 
```

```python
#3) f.write(string) writes the contents of string to the file, 
#   returning the number of characters written

with open('new_file','w') as f:
  print(f.write('this shows how this function works'))    #returns the number of characters in the string

with open('new_file','r') as f:
  read_file = f.read()
print(read_file)
```

```python
#4) writeline()
lines='''
  India is a glorious country.
  I am proud to be an Indian.
  It has a properous history also.
'''

with open('new_file2','w+') as f:
  f.writelines(lines)
  f.seek(0)   #to reset the pointer to the first position in the file
  read_data = f.read()
print(read_data)
```

```python

f = open('new_file','a+')

f.write("I am a good boy")
read_data = f.read()
print(read_data)  #this gives an empty line because in append mode file pointer is at the end
f.close()
print(f.closed)
```

```python
#to specify binary or text mode we use t or b subsript
f = open('new_file','rb')
read_data = f.read()
print(read_data)
```

```python
#4) f.tell()
# we want to know where is our file pointer write now
with open('new_file2','a+') as f:
  f.write('\nWe here a lot of inspiring stories in India')
  print(f.tell())   #it tells that the pointer is at last now we can use seek to return it to the starting
  f.seek(0)
  print(f.tell())
  read_data = f.read()
print(read_data)
```

```python
#5) f.seek()
# it is used to change the file pointer--
with open('new_file2','a+') as f:
  f.seek(3)     #go to the third byte (0 based indexing is followed it can be negative also llike strings)
  read_data=f.read(5) #read next 5 bytes
print(read_data)

#second argument of f.seek()--
#offset
# 0 => set pointer from starting
# 1 => set pointer from current position
# 2 => set pointer from end of the file

```

```python
#6) truncate method--
# f.truncate(5) only 5 bytes will remain in the file

with open('new_file3','w+') as f:
  f.write("My name is Bhanu Verma")
  f.truncate(10)
  f.seek(0)
  read_data = f.read()

print(read_data)
```
